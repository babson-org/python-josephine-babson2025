    The Monty Hall choice:

    You are presented with three closed doors. A car is behind one,   
    and goats are behind the other two.  You pick one door (e.g., Door 1).  
    The host, who knows where the car is, opens one of the other two     
    doors (e.g., Door 3) to reveal a goat. The host then asks if you want   
    to stick with your original choice (Door 1) or switch to the other  
    remaining unopened door (Door 2).   

    Create simulations where you test both strategies: staying with your  
    original guess or switching to the other door.   

    For instance: suppose you choose door 1 (and door 1 happens to have   
    the car behind it).  Monty reveals either  door 2 or door 3  
    (assume he selects one with equal probabilities).  In the no switching    
    simulation you keep door 1 and win the car. In the switching strategy  
    you pick the door he doesn't show (2 or 3) and get the goat.

```python
'''
(enlarge sections as needed)
1) Computational thinking 7 1/2 points:
    -doors are integers (1,2,3)
    -strategies are booleans (false-no switch, true - switch)
    -get number of trials
    - loop through switching and not switching
        -prizes are random
        -monty reveals goat first
        -switch or stay
        - track wins and number of times each door is revealed
    -display results as a percentage











2) Pseudo code 7 1/2 points:

ask user for numer of trials
    validate positive input

switching vs no switch - [false, true]

for each switch
    add to wins
    see reveals

    do for all trials
        doors = [1,2,3]
        prize = random door
        contestant = random door

        reveal = not prize and not contestant
        
        if they switch
            open new door

        if they win prize
            add to win total

    print the stats











3) Actual code 10 points:
'''

#You don't need to use this structure but consider the following:
import random
def monty_sim():
    txt = 'Enter the number of simulations you want to run'
    
    while True:
        try:
            trials = int(input(txt))
            if trials > 0:
                break
            else:
                print("Please enter a positive integer.")
        except ValueError:
            print("Invalid input. Please enter a number.")




    strategy = [False, True]

    
    for switch in strategy:                         # loop to run both strategies
        wins = 0                                    # initialize wins for each strategy
        reveals = [0, 0, 0]                         # initialize reveals for each strategy i.e the number of times Monty reveals a given door
        for _ in range(trials):                     # loop to run simulations
            doors = [1, 2, 3]                       # for each trial we reset doors. We can eliminate both the prize door and the contestant's initial choice to see what's left
            prize = random.choice(doors)  
            contestant = random.choice(doors)

            # Monty reveals a goat door (not switching)
            can_reveal = [
                door for door in doors
                if door != prize and door != contestant
            ]
            reveal = random.choice(can_reveal)
            reveals[reveal - 1] += 1               # which door has monty revealed

            # Switching doors
            if switch:
                for door in doors:
                    if door != contestant and door != reveal:
                        contestant = door           # swtiched to door that can't be the previous one or the one thats been revealed
                        break
                
            # Check win
            if contestant == prize:
                wins += 1
        

    #SUGGESTION: first code the no switch strategy i.e. switch = False, then use switch to add addional logic when switch = True









        label = "WITH Switching" if switch else "NO Switching"
        print(f"\n--- {label} ---")
        print(f"Winning percentage: {wins / trials:.2%}")
        print(f"Losing percentage:  {1 - wins / trials:.2%}")
        print(f"Monty revealed door 1: {reveals[0]:,}")
        print(f"                    2: {reveals[1]:,}")
        print(f"                    3: {reveals[2]:,}\n")
        print(f" revealed doors sum to {sum(reveals):,}")




monty_sim()
```
